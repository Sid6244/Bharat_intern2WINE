```python
#importing required libraries
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn import datasets
from sklearn.metrics import classification_report
```

```python
#loading the wine dataset
wine = datasets.load_wine() 
```

```python
#arranging the dataset into pandas dataframe
df = pd.DataFrame(wine['data'], columns = wine['feature_names']) 
df
```

```python
 #adding target column in the dataframe
df['target'] = wine['target'] 
df
```

```python
#checking for null values in the dataframe
df.isnull().sum() 
```

```python
 #checking the statistical measures of the dataframe
df.describe()
```

```python
df.info()
```

```python
#Splitting into Features(X) and Target(Y)
X = df
Y=wine.target
Y
```

```python
print("X - Shape : ", X.shape)
print("Y - shape : ", Y.shape)
```

```python
#Data Segregation, 80% (X_train, Y_train) 20% (X_test, Y_test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 45)
```

```python
#Training the model 
l_model = LinearRegression()
l_model.fit(X_train,Y_train)
```

```python
#Making predictions on Testing data
Val_pred=l_model.predict(X_test)
Val_pred
```

```python
#Checking the accuracy of the model using r2_score
score1=metrics.r2_score(Y_test,Val_pred)
score1
```

```python
#Visualising the actual and predicted prices on VALIDATION DATA 
plt.scatter(Y_test,Val_pred) 
plt.xlabel("Actual Quality") 
plt.ylabel("Predicted Quality")
plt.title(" Actual Quality vs Predicted Quality")
plt.show()
```
